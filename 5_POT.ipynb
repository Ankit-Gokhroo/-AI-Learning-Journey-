```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
nltk.download('punkt_tab')
```

```python
stemmer = PorterStemmer()
```

```python
nltk.download('stopwords')
```

```python
print(stopwords.words('english'))
```

```python
corpus = """My dear young friends, I stand before you today, not just as a former President, but as a fellow dreamer—someone who believes that each one of you has the power to transform India. Dreams, you know, are not those which come while you are sleeping; rather, dreams are those that do not let you sleep! When I was a small boy in Rameswaram, I often wondered, ‘Can I really make a difference?’ The answer, I discovered, lies within ourselves. You must ask: what will I give to the nation? Not ‘what will the nation give me?’ because true success doesn’t come from taking—it comes from giving, striving, and persisting.

Education, dedication, and hard work—these are the keys; however, one must also have courage, curiosity, and compassion. Sometimes, you may fail (and that’s okay!), but remember: failure is not the opposite of success; it’s a part of success. If you learn from it, you grow. So, don’t fear failure—embrace it, analyze it, and move forward.

Let your vision be big, your actions be consistent, and your intentions be pure. Together, we can build a developed India—an India that is strong, self-reliant, and full of innovation. The journey may be long, the road may be tough, but with determination, discipline, and dreams, nothing is impossible.
"""
```

```python
sentences = nltk.sent_tokenize(corpus)
sentences
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# we will find the POS tag

for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  # sentences[i] = ' '.join(words) #converting all the list words into sentences
  pos_tag = nltk.pos_tag(words)
  print(pos_tag)
```
