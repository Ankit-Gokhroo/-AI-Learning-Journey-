```python
import nltk
```

```python
words = ["eating", "eats", "eaten", "writing", "writes", "programming", "programs", "history", "finally", "finalized"]
```

```python
#porterStemmer
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for word in words:
  print(word+"------>>"+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('ssit') #few cons
```

```python
#RegexpStemmer -- takes single re and removes any prefix or suffix that matches expression
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)

reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('eatinging')
```

```python
#SnowballStemmer --better accuracy compare to porterStemmer but still cons
from nltk.stem import SnowballStemmer
snowBallStem = SnowballStemmer('english')
```

```python
for word in words:
  print(word+"------>>"+snowBallStem.stem(word))
```

```python
stemming.stem('fairly'), stemming.stem('sportingly')
```

```python
snowBallStem.stem('fairly'), snowBallStem.stem('sportingly')
```
