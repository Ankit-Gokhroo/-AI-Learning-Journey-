```python
import nltk

#tokenization

## paragraph >> sentence
from nltk.tokenize import sent_tokenize

## paragraph >> words, sentence >> words
from nltk.tokenize import word_tokenize

from nltk.tokenize import wordpunct_tokenize
from nltk.tokenize import TreebankWordTokenizer
```

```python
#nltk.download('punkt_tab')
```

```python
corpus = """Hello Welcome, to Ankit Gokhroo Profile's here.
Connect with me, am open for discussion, thank you ! for reach out.
"""
print(corpus)
```

```python
documents = sent_tokenize(corpus) # seperate by . and !
documents
```

```python
print(word_tokenize(corpus)) # sperated {, . '} 'S not splitted
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
print(wordpunct_tokenize(corpus)) # 's also splitted
```

```python
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus) #last full stop  only considered sepearte
```
